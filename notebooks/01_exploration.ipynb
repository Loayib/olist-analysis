# Olist E-Commerce Analysis
## Notebook 1 — Data Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

DATA = 'C:/Users/loayi/OneDrive/Desktop/Workspace/olist-analysis/data/'

customers    = pd.read_csv(DATA + 'olist_customers_dataset.csv')
orders       = pd.read_csv(DATA + 'olist_orders_dataset.csv')
order_items  = pd.read_csv(DATA + 'olist_order_items_dataset.csv')
payments     = pd.read_csv(DATA + 'olist_order_payments_dataset.csv')
reviews      = pd.read_csv(DATA + 'olist_order_reviews_dataset.csv')
products     = pd.read_csv(DATA + 'olist_products_dataset.csv')
sellers      = pd.read_csv(DATA + 'olist_sellers_dataset.csv')
translations = pd.read_csv(DATA + 'product_category_name_translation.csv')

tables = {'customers': customers, 'orders': orders, 'order_items': order_items,
          'payments': payments, 'reviews': reviews, 'products': products,
          'sellers': sellers, 'translations': translations}

for name, df in tables.items():
    print(f'{name}: {df.shape}')
```

## Missing Values

```python
for name, df in tables.items():
    nulls = df.isnull().sum()
    nulls = nulls[nulls > 0]
    if len(nulls) > 0:
        print(f'\n{name}:')
        print(nulls)
```

## Order Status Distribution

```python
orders['order_status'].value_counts()
```

## Orders Over Time

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['month'] = orders['order_purchase_timestamp'].dt.to_period('M')
monthly = orders.groupby('month')['order_id'].count()

monthly.plot(figsize=(14, 5), marker='o')
plt.title('Number of Orders per Month')
plt.xlabel('Month')
plt.ylabel('Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Insight:** Order volume grew steadily from 2016 to a peak in late 2017/early 2018, consistent with Black Friday and holiday season demand. The sharp decline at the end reflects the dataset cutoff, not an actual drop in sales.

## Top 10 Product Categories

```python
products_translated = pd.merge(products, translations, on='product_category_name', how='left')
top_cats = products_translated['product_category_name_english'].value_counts().head(10)

top_cats.plot(kind='barh', figsize=(10, 6))
plt.title('Top 10 Product Categories')
plt.xlabel('Number of Products')
plt.tight_layout()
plt.show()
```

**Insight:** Top categories are Bed/Bath, Sports, and Furniture — Brazilians on Olist buy mostly home goods and lifestyle products.

## Review Score Distribution

```python
reviews['review_score'].value_counts().sort_index().plot(kind='bar', figsize=(8, 4))
plt.title('Review Score Distribution')
plt.xlabel('Score')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

**Insight:** Reviews are polarized — most customers leave a 5 with a secondary spike at 1. Middle scores are rare, which is typical in e-commerce.
